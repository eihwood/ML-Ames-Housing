```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression
ridge = Ridge()
lasso = Lasso()
net   = ElasticNet()
```

```python
# Import Austin's very nicely cleaned data :) 

data = pd.read_csv('../Ames_HousePrice_processed.csv')
```

```python
price = data['SalePrice']
features = data.drop(['SalePrice', 'PID'], axis = 1)
feature_dummified = pd.get_dummies(features, drop_first = True)
feature_dummified.columns.to_list()
```

```python
lasso.set_params(alpha=1, normalize=True)

```

```python
lasso.fit(feature_dummified, price)
print('The intercept is %.4f' %(lasso.intercept_))
lassoCoef = pd.Series(lasso.coef_, index=feature_dummified.columns)
print('The slopes are \n%s' %(lassoCoef))
```

```python
coefs = []
intercepts = []
R2  = []
alphaRange = np.linspace(1e-3,20,20)
for alpha in alphaRange:
    lasso.set_params(alpha=alpha, normalize = True)  
    lasso.fit(feature_dummified, price)
    intercepts.append(lasso.intercept_)
    coefs.append(lasso.coef_)
    R2.append(lasso.score(feature_dummified, price))  
```

```python
coefs = pd.DataFrame(np.array(coefs), columns=feature_dummified.columns)

for name in coefs.columns:
    plt.plot(alphaRange, coefs[name], label=name)
plt.legend(loc=1)    
plt.title('Change of Lasso Slopes Varying Alpha')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature_dummified, price, test_size=0.33, random_state=42)

```

```python
# TRY MLR

import statsmodels.api as sm

X_add_const = sm.add_constant(X_train)
ols = sm.OLS(y_train, X_add_const)
ans = ols.fit()
print(ans.summary())

res = pd.DataFrame(ans.summary().tables[1])

res.sort()


```
